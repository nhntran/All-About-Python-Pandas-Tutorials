### 1. BASIC PLOTS

```python
df_titanic.hist(figsize = (10,8)); #Changing figure size to see better
```

```python
df_titanic.plot(kind = 'box', figsize = (10, 8));
```

```python
## Plot seperate variable
df_titanic['Age'].hist()

## Plot using general function 'plot'
df_titanic['Age'].plot(kind = 'hist');
```

```python
## Using value_counts() to plot class variable
df_titanic['Embarked'].value_counts().plot(kind = 'bar');
```

```python
## Using value_counts() to plot class variable - Using pie chart
df_titanic['Embarked'].value_counts().plot(kind = 'pie', figsize = (8,8));
```

```python
## Scatter plot, relationship between 2 variabels
df_titanic.plot(x = 'Age', y = 'Survived', kind = 'scatter');
```

```python
## Box plot
df_titanic['Age'].plot(kind = 'box');
```

```python
## 2 box plots belonging to 2 different dataframes into a figure
df_plot1 = pd.DataFrame(np.random.rand(3,2), columns = ['First', 'Second'])
df_plot2 = pd.DataFrame(np.random.rand(3,2)*5, columns = ['First', 'Second'])

## Drawing box plot
ax = df_plot1.plot(kind='box',
             color=dict(boxes='r', whiskers='r', medians='r', caps='r'),
             boxprops=dict(linestyle='-', linewidth=1.5),
             flierprops=dict(linestyle='-', linewidth=1.5),
             medianprops=dict(linestyle='-', linewidth=1.5),
             whiskerprops=dict(linestyle='-', linewidth=1.5),
             capprops=dict(linestyle='-', linewidth=1.5),
             showfliers=False, grid=True, rot=0, figsize = (8, 6))

df_plot2.boxplot(ax = ax)
```
